### (1.2.1) XGBoost-BOA: Convergence Analysis

```python
# --- Bayesian Optimization with Convergence Diagnostics & Early Stopping ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import erf, sqrt, exp, pi
from bayes_opt import BayesianOptimization

# =========================
# Objective (your predictor)
# =========================
def predict_xgboost(FS, TWAS):
    """
    Wraps your trained model as a black-box function for BO to maximize.
    Expects a global 'regressor' with .predict(pd.DataFrame) -> [float].
    """
    data = pd.DataFrame({'FS': FS, 'TWAS': TWAS},  index=[0])
    y = float(regressor.predict(data)[0])
    # Guardrails in case model yields NaN/inf
    if not np.isfinite(y):
        return -1e15  # penalize invalid regions
    return y

# =========================
# Bounds
# =========================
pbounds = {'FS': (0.53, 8.94), 'TWAS': (2.4, 11.03)}

# =========================
# BO setup
# =========================
optimizer = BayesianOptimization(
    f=predict_xgboost,
    pbounds=pbounds,
    random_state=42,
    verbose=2
)

# =========================
# Convergence Criteria (tunable)
# =========================
MAX_ITERS       = 1500   # hard cap
INIT_POINTS     = 10      # random initial evals

WINDOW          = 400     # best-so-far stabilization window (iterations)
IMPROVEMENT_EPS = 1e-4   # absolute gain threshold over WINDOW to consider "stabilized"

GRID_SIZE       = 3000   # random grid points to probe acquisition & GP uncertainty
ACQ_XI          = 0.5   # exploration parameter for EI
ACQ_FLAT_EPS    = 1e-07   # if (max_acq - min_acq) < ACQ_FLAT_EPS => acquisition is "flat"

POST_STD_EPS    = 5e-3   # mean predictive std threshold over grid

# =========================
# Initialize (random points only)
# =========================
optimizer.maximize(init_points=INIT_POINTS, n_iter=0)

# =========================
# Helpers
# =========================
def _random_grid(n, bounds):
    keys = list(bounds.keys())
    X = np.column_stack([
        np.random.uniform(low=bounds[k][0], high=bounds[k][1], size=n)
        for k in keys
    ])
    return X

# Normal PDF/CDF (no scipy dependency)
SQRT_2PI = sqrt(2.0 * pi)
def _norm_pdf(z):
    z = np.asarray(z, dtype=float)
    return np.exp(-0.5 * z * z) / np.sqrt(2.0 * np.pi)

def _norm_cdf(x):
    """
    Abramowitz & Stegun 7.1.26 approximation for Î¦(x), vectorized, no scipy/erf.
    Max abs error ~ 7.5e-8, good for BO needs.
    """
    x = np.asarray(x, dtype=float)
    t = 1.0 / (1.0 + 0.2316419 * np.abs(x))
    # Polynomial in t
    poly = (((((1.330274429 * t) + (-1.821255978)) * t + 1.781477937) * t
             + (-0.356563782)) * t + 0.319381530) * t
    pdf = _norm_pdf(x)
    cdf_pos = 1.0 - pdf * poly
    # Reflect for negative x
    return np.where(x >= 0.0, cdf_pos, 1.0 - cdf_pos)

def expected_improvement(mu, sigma, y_max, xi):
    """
    Vectorized EI using the A&S CDF approximation; safe for sigma=0.
    """
    mu = np.asarray(mu, dtype=float).ravel()
    sigma = np.asarray(sigma, dtype=float).ravel()
    imp = mu - y_max - xi

    ei = np.zeros_like(mu)
    nonzero = sigma > 0.0
    Z = np.zeros_like(mu)
    Z[nonzero] = imp[nonzero] / sigma[nonzero]

    Phi = _norm_cdf(Z[nonzero])
    phi = _norm_pdf(Z[nonzero])
    ei[nonzero] = imp[nonzero] * Phi + sigma[nonzero] * phi
    ei[~nonzero] = np.maximum(0.0, imp[~nonzero])  # when sigma == 0

    return ei

# Logging containers
history = {
    "iter": [],
    "target": [],
    "best_so_far": [],
    "acq_span": [],
    "mean_post_std": [],
    "stop_reason": []
}

stop_reason = None

# =========================
# Iterative optimization with diagnostics
# =========================
for _ in range(1, MAX_ITERS + 1):
    # One optimization step
    optimizer.maximize(init_points=0, n_iter=1)  # acquisition handled internally by bayes_opt

    # Current best and last evaluation
    best_val = optimizer.max["target"]
    last_val = optimizer.res[-1]["target"]

    # Probe acquisition & posterior uncertainty on a random grid
    X_grid = _random_grid(GRID_SIZE, pbounds)
    gp = optimizer._gp  # underlying GaussianProcessRegressor

    # Predict GP mean/std on the grid
    mu, stds = gp.predict(X_grid, return_std=True)
    mean_post_std = float(np.mean(stds))

    # Compute EI (Expected Improvement) ourselves to assess "acquisition flatness"
    y_max = optimizer._space.target.max() if len(optimizer._space.target) else -np.inf
    acq_vals = expected_improvement(mu, stds, y_max=y_max, xi=ACQ_XI)
    acq_span = float(np.max(acq_vals) - np.min(acq_vals))

    # Log
    history["iter"].append(len(optimizer.res))
    history["target"].append(last_val)
    history["best_so_far"].append(best_val)
    history["acq_span"].append(acq_span)
    history["mean_post_std"].append(mean_post_std)
    history["stop_reason"].append("")

    # ---- Convergence checks ----
    # 1) Best-so-far stabilization
    if len(history["best_so_far"]) > WINDOW:
        window_best = history["best_so_far"][-WINDOW:]
        improvement = window_best[-1] - window_best[0]
        if improvement < IMPROVEMENT_EPS:
            stop_reason = f"Best-so-far stabilized (< {IMPROVEMENT_EPS} over last {WINDOW} iters)"

    # 2) Acquisition flatness (via EI span)
    if stop_reason is None and acq_span < ACQ_FLAT_EPS:
        stop_reason = f"Acquisition flat (EI span {acq_span:.2e} < {ACQ_FLAT_EPS})"

    # 3) Posterior uncertainty small
    if stop_reason is None and mean_post_std < POST_STD_EPS:
        stop_reason = f"Posterior uncertainty small (mean Ïƒ {mean_post_std:.2e} < {POST_STD_EPS})"

    # Break if any criterion triggered
    if stop_reason is not None:
        history["stop_reason"][-1] = stop_reason
        break

# =========================
# Report & Plots
# =========================
best_params = optimizer.max["params"]
best_target = optimizer.max["target"]

print("==== Convergence Report ====")
print(f"Iterations run: {len(optimizer.res)} / {MAX_ITERS}")
print(f"Best target: {best_target:.6f}")
print(f"Best params: {best_params}")
print(f"Stop reason: {stop_reason or 'Hit MAX_ITERS'}")

# Convert logs to DataFrame (optional)
log_df = pd.DataFrame(history)

# --- Plot: Best-so-far trace ---
plt.figure()
plt.plot(log_df["iter"], log_df["best_so_far"])
plt.xlabel("Iteration")
plt.ylabel("Best-so-far target")
plt.title("Best-so-far convergence trace")
plt.grid(True, alpha=0.5)
plt.tight_layout()
plt.show()

# --- Plot: Acquisition flatness diagnostic (EI span) ---
plt.figure()
plt.plot(log_df["iter"], log_df["acq_span"])
plt.xlabel("Iteration")
plt.ylabel("Acquisition span (EI max - min)")
plt.title("Acquisition flatness diagnostic")
plt.grid(True, alpha=0.5)
plt.tight_layout()
plt.show()

# --- Plot: Posterior uncertainty diagnostic ---
plt.figure()
plt.plot(log_df["iter"], log_df["mean_post_std"])
plt.xlabel("Iteration")
plt.ylabel("Mean posterior std (Ïƒ)")
plt.title("Posterior uncertainty diagnostic")
plt.grid(True, alpha=0.5)
plt.tight_layout()
plt.show()
```

```python

merged_Biogas1_3=merged_Biogas1_2.copy()

def remove_outliers_mad(data, column, multiplier=3.5):
    median = data[column].median()
    mad = (data[column] - median).abs().median()
    
    # The constant 1.4826 makes MAD an estimator of the standard deviation 
    # assuming the data follows a normal distribution.
    mad_adjusted = mad * 1.4826
    
    # Define bounds for outliers
    lower_bound = median - multiplier * mad_adjusted
    upper_bound = median + multiplier * mad_adjusted

    # Filtering data without copying DataFrame
    initial_count = data.shape[0]
    data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    final_count = data.shape[0]

    print(f'Removed {initial_count - final_count} outliers from {column} using MAD')

    return data

# Apply the function for each column
columns_to_remove_outliers = ['TWAS_flow', 'pH_FS','VS_FS','T_FS','FS_flow','Biogas_flow']
for column in columns_to_remove_outliers:
    merged_Biogas1_3 = remove_outliers_mad(merged_Biogas1_3, column, multiplier=3.5)

# Optionally, reset the index after removing outliers
merged_Biogas1_3 = merged_Biogas1_3.reset_index(drop=True)

# Display or return the cleaned DataFrame
merged_Biogas1_3
```

## XGBoost-BOA Convergence evaluations

```python
# --- Bayesian Optimization with Convergence Diagnostics & Early Stopping ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import erf, sqrt, exp, pi
from bayes_opt import BayesianOptimization

# =========================
# Objective (your predictor)
# =========================
def predict_xgboost(TWAS_flow, pH_FS, VS_FS, T_FS, FS_flow):
    """
    Wraps your trained model as a black-box function for BO to maximize.
    Expects a global 'regressor' with .predict(pd.DataFrame) -> [float].
    """
    data = pd.DataFrame({
        'TWAS_flow': [TWAS_flow],
        'pH_FS':     [pH_FS],
        'VS_FS':     [VS_FS],
        'T_FS':      [T_FS],
        'FS_flow':   [FS_flow],
    })
    y = float(regressor2.predict(data)[0])
    # Guardrails in case model yields NaN/inf
    if not np.isfinite(y):
        return -1e15  # penalize invalid regions
    return y

# =========================
# Bounds
# =========================
pbounds = {
    'TWAS_flow': (2.4, 11.03),
    'pH_FS':     (4.75, 6.7),
    'VS_FS':     (0.34, 6.9),
    'T_FS':      (11.95, 21.7),
    'FS_flow':   (1.0, 8.2),
}

# =========================
# BO setup
# =========================
optimizer = BayesianOptimization(
    f=predict_xgboost,
    pbounds=pbounds,
    random_state=42,
    verbose=2
)


# =========================
# Convergence Criteria (tunable)
# =========================
MAX_ITERS       = 1500   # hard cap
INIT_POINTS     = 100      # random initial evals

WINDOW          = 100     # best-so-far stabilization window (iterations)
IMPROVEMENT_EPS = 1e-5   # absolute gain threshold over WINDOW to consider "stabilized"

GRID_SIZE       = 4000   # random grid points to probe acquisition & GP uncertainty
ACQ_XI          = 0.05   # exploration parameter for EI
ACQ_FLAT_EPS    = 1e-6   # if (max_acq - min_acq) < ACQ_FLAT_EPS => acquisition is "flat"

POST_STD_EPS    = 2e-2   # mean predictive std threshold over grid

# =========================
# Initialize (random points only)
# =========================
optimizer.maximize(init_points=INIT_POINTS, n_iter=0)

# =========================
# Helpers
# =========================
def _random_grid(n, bounds):
    keys = list(bounds.keys())
    X = np.column_stack([
        np.random.uniform(low=bounds[k][0], high=bounds[k][1], size=n)
        for k in keys
    ])
    return X

# Normal PDF/CDF (no scipy dependency)
SQRT_2PI = sqrt(2.0 * pi)
def _norm_pdf(z):
    z = np.asarray(z, dtype=float)
    return np.exp(-0.5 * z * z) / np.sqrt(2.0 * np.pi)

def _norm_cdf(x):
    """
    Abramowitz & Stegun 7.1.26 approximation for Î¦(x), vectorized, no scipy/erf.
    Max abs error ~ 7.5e-8, good for BO needs.
    """
    x = np.asarray(x, dtype=float)
    t = 1.0 / (1.0 + 0.2316419 * np.abs(x))
    # Polynomial in t
    poly = (((((1.330274429 * t) + (-1.821255978)) * t + 1.781477937) * t
             + (-0.356563782)) * t + 0.319381530) * t
    pdf = _norm_pdf(x)
    cdf_pos = 1.0 - pdf * poly
    # Reflect for negative x
    return np.where(x >= 0.0, cdf_pos, 1.0 - cdf_pos)

def expected_improvement(mu, sigma, y_max, xi):
    """
    Vectorized EI using the A&S CDF approximation; safe for sigma=0.
    """
    mu = np.asarray(mu, dtype=float).ravel()
    sigma = np.asarray(sigma, dtype=float).ravel()
    imp = mu - y_max - xi

    ei = np.zeros_like(mu)
    nonzero = sigma > 0.0
    Z = np.zeros_like(mu)
    Z[nonzero] = imp[nonzero] / sigma[nonzero]

    Phi = _norm_cdf(Z[nonzero])
    phi = _norm_pdf(Z[nonzero])
    ei[nonzero] = imp[nonzero] * Phi + sigma[nonzero] * phi
    ei[~nonzero] = np.maximum(0.0, imp[~nonzero])  # when sigma == 0

    return ei

# Logging containers
history = {
    "iter": [],
    "target": [],
    "best_so_far": [],
    "acq_span": [],
    "mean_post_std": [],
    "stop_reason": []
}

stop_reason = None

# =========================
# Iterative optimization with diagnostics
# =========================
for _ in range(1, MAX_ITERS + 1):
    # One optimization step
    optimizer.maximize(init_points=0, n_iter=1)  # acquisition handled internally by bayes_opt

    # Current best and last evaluation
    best_val = optimizer.max["target"]
    last_val = optimizer.res[-1]["target"]

    # Probe acquisition & posterior uncertainty on a random grid
    X_grid = _random_grid(GRID_SIZE, pbounds)
    gp = optimizer._gp  # underlying GaussianProcessRegressor

    # Predict GP mean/std on the grid
    mu, stds = gp.predict(X_grid, return_std=True)
    mean_post_std = float(np.mean(stds))

    # Compute EI (Expected Improvement) ourselves to assess "acquisition flatness"
    y_max = optimizer._space.target.max() if len(optimizer._space.target) else -np.inf
    acq_vals = expected_improvement(mu, stds, y_max=y_max, xi=ACQ_XI)
    acq_span = float(np.max(acq_vals) - np.min(acq_vals))

    # Log
    history["iter"].append(len(optimizer.res))
    history["target"].append(last_val)
    history["best_so_far"].append(best_val)
    history["acq_span"].append(acq_span)
    history["mean_post_std"].append(mean_post_std)
    history["stop_reason"].append("")

    # ---- Convergence checks ----
    # 1) Best-so-far stabilization
    if len(history["best_so_far"]) > WINDOW:
        window_best = history["best_so_far"][-WINDOW:]
        improvement = window_best[-1] - window_best[0]
        if improvement < IMPROVEMENT_EPS:
            stop_reason = f"Best-so-far stabilized (< {IMPROVEMENT_EPS} over last {WINDOW} iters)"

    # 2) Acquisition flatness (via EI span)
    if stop_reason is None and acq_span < ACQ_FLAT_EPS:
        stop_reason = f"Acquisition flat (EI span {acq_span:.2e} < {ACQ_FLAT_EPS})"

    # 3) Posterior uncertainty small
    if stop_reason is None and mean_post_std < POST_STD_EPS:
        stop_reason = f"Posterior uncertainty small (mean Ïƒ {mean_post_std:.2e} < {POST_STD_EPS})"

    # Break if any criterion triggered
    if stop_reason is not None:
        history["stop_reason"][-1] = stop_reason
        break

# =========================
# Report & Plots
# =========================
best_params = optimizer.max["params"]
best_target = optimizer.max["target"]

print("==== Convergence Report ====")
print(f"Iterations run: {len(optimizer.res)} / {MAX_ITERS}")
print(f"Best target: {best_target:.6f}")
print(f"Best params: {best_params}")
print(f"Stop reason: {stop_reason or 'Hit MAX_ITERS'}")

# Convert logs to DataFrame (optional)
log_df = pd.DataFrame(history)

# --- Plot: Best-so-far trace ---
plt.figure()
plt.plot(log_df["iter"], log_df["best_so_far"])
plt.xlabel("Iteration")
plt.ylabel("Best-so-far target")
plt.title("Best-so-far convergence trace")
plt.grid(True, alpha=0.5)
plt.tight_layout()
plt.show()

# --- Plot: Acquisition flatness diagnostic (EI span) ---
plt.figure()
plt.plot(log_df["iter"], log_df["acq_span"])
plt.xlabel("Iteration")
plt.ylabel("Acquisition span (EI max - min)")
plt.title("Acquisition flatness diagnostic")
plt.grid(True, alpha=0.5)
plt.tight_layout()
plt.show()

# --- Plot: Posterior uncertainty diagnostic ---
plt.figure()
plt.plot(log_df["iter"], log_df["mean_post_std"])
plt.xlabel("Iteration")
plt.ylabel("Mean posterior std (Ïƒ)")
plt.title("Posterior uncertainty diagnostic")
plt.grid(True, alpha=0.5)
plt.tight_layout()
plt.show()
```
